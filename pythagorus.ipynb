```python
from math import sqrt

```

```python
n=int(input("enter n")) 
```

```python
for a in range (1,n+1):
    for b in range (a,n):
        print ( a,b)
        
      
        c_sq=a**2+b**2
        c=int(sqrt(c_sq))
        #print ("c is ", c)
        if ((c_sq-c**2)==0):
            print(a,b,c)
    
```
